```python
import cv2
```

```python
import numpy as np
file_path = "/content/Untitled/"
```

```python
yolo_weights = file_path+"yolov3.weights" # 가중치
yolo_config = file_path+ "yolov3.cfg"# 설정
coco_names = file_path+"coco.names"#이름들
```

```python
with open(coco_names, 'r') as f:
    classes = [line.strip() for line in f.readlines()]
classes
```

```python
from google.colab.patches import cv2_imshow
```

```python
with open(coco_names, 'r') as f:
    classes = [line.strip() for line in f.readlines()]


net = cv2.dnn.readNet(yolo_weights, yolo_config)
image = cv2.imread("/content/KakaoTalk_20241021_154940431.jpg")
height , width = image.shape[:2]
#이미지 전처리
blob = cv2.dnn.blobFromImage(image, 1/255.0, (416, 416), swapRB=True, crop=False)
net.setInput(blob)
layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]
# 객체 ㅁ지
outs = net.forward(output_layers)
#$ 탐지된객체 정보 저장
class_ids = []
confidences = []
boxes = []
for out in outs:
    for detection in out:
        scores = detection[5:]
        class_id = np.argmax(scores)
        confidence =  scores[class_id]
        if confidence > 0.5:  # 신뢰도 0.5 이상인 경우
            center_x = int(detection[0] * width)  # 중심 x 좌표
            center_y = int(detection[1] * height)  # 중심 y 좌표
            w = int(detection[2] * width)  # 객체의 너비
            h = int(detection[3] * height)  # 객체의 높이

            # 경계 상자의 좌상단 좌표 계산
            x = int(center_x - w / 2)
            y = int(center_y - h / 2)

            # 탐지된 객체 정보 저장
            boxes.append([x, y, w, h])  # 경계 상자 정보
            confidences.append(float(confidence))  # 신뢰도 점수
            class_ids.append(class_id)  # 클래스 ID

# 비최대 억제 (Non-Maximum Suppression)
indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

# 결과 시각화
for i in range(len(boxes)):
    if i in indexes:
        x, y, w, h = boxes[i]
        label = str(classes[class_ids[i]])
        confidence = confidences[i]
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
        cv2.putText(image, f'{label} {confidence:.2f}', (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

# 결과 이미지 출력
cv2_imshow(image)

```
